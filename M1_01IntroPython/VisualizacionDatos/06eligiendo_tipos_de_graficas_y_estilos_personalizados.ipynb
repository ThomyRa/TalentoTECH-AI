Dado que no siempre es fácil decidir cuál es la mejor forma de contar la historia detrás de tus datos, hemos dividido los tipos de gráficos en tres categorías generales para ayudarte con esto.

- **Tendencias** – Una tendencia se define como un patrón de cambio.
    - `sns.lineplot` – Los **gráficos de líneas** son ideales para mostrar tendencias a lo largo del tiempo, y se pueden usar múltiples líneas para mostrar tendencias en más de un grupo.

- **Relaciones** – Existen muchos tipos de gráficos que puedes utilizar para entender las relaciones entre variables en tus datos.
    - `sns.barplot` – Los **gráficos de barras** son útiles para comparar cantidades entre diferentes grupos.
    - `sns.heatmap` – Los **mapas de calor** permiten identificar patrones codificados por color en tablas de números.
    - `sns.scatterplot` – Los **diagramas de dispersión** muestran la relación entre dos variables continuas; si se codifican por color, también se puede representar la relación con una tercera [variable categórica](https://es.wikipedia.org/wiki/Variable_categórica).
    - `sns.regplot` – Incluir una **línea de regresión** en un diagrama de dispersión facilita la visualización de una posible relación lineal entre dos variables.
    - `sns.lmplot` – Este comando es útil para trazar múltiples líneas de regresión si el diagrama de dispersión contiene varios grupos codificados por color.
    - `sns.swarmplot` – Los **diagramas de dispersión categóricos** muestran la relación entre una variable continua y una variable categórica.

- **Distribución** – Visualizamos distribuciones para mostrar los valores posibles que puede tomar una variable, junto con su probabilidad.
    - `sns.histplot` – Los **histogramas** muestran la distribución de una sola variable numérica.
    - `sns.kdeplot` – Los **gráficos KDE** (o **KDE 2D**) muestran una distribución suavizada estimada de una o dos variables numéricas.
    - `sns.jointplot` – Este comando es útil para mostrar simultáneamente un gráfico KDE 2D junto con los gráficos KDE correspondientes para cada variable individual.

# Cambiando estilos con seaborn

Todos los comandos anteriores proporcionan un estilo predeterminado agradable para cada uno de los gráficos. Sin embargo, puede que te resulte útil personalizar el aspecto de tus gráficos, y por suerte, ¡esto puede lograrse simplemente añadiendo una línea de código más!

Como siempre, debemos comenzar configurando el entorno de programación.

```python

import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

Trabajaremos con el mismo código que utilizamos para crear un gráfico de líneas en una seccion anterior. El siguiente código carga el conjunto de datos y crea el gráfico.



```python
# Path of the file to read
spotify_filepath = "/content/spotify.csv"

# Read the file into a variable spotify_data
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)

# Line chart
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```

Podemos cambiar rápidamente el estilo de la figura a un tema diferente con solo una línea de código.


```python
# Change the style of the figure to the "dark" theme
sns.set_style("dark")

# Line chart
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```

Seaborn tiene cinco temas diferentes: (1)`"darkgrid"`, (2)`"whitegrid"`, (3)`"dark"`, (4)`"white"` y (5)`"ticks"`, y solo necesitas usar un comando similar al de la celda de código anterior (rellenando con el tema elegido) para cambiarlo.
