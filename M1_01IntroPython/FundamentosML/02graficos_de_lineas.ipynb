# Graficar los datos

Ahora que el conjunto de datos está cargado en el notebook, ¡solo necesitamos una línea de código para crear un gráfico de líneas!


````python
# Line chart showing daily global streams of each song
sns.lineplot(data=spotify_data)
````

Como puedes ver arriba, la línea de código es relativamente corta y tiene dos componentes principales:
- `sns.lineplot` le indica al notebook que queremos crear un gráfico de líneas.  
  - _Cada comando que aprendas en este curso comenzará con `sns`, lo que indica que el comando proviene de la librería [seaborn](https://seaborn.pydata.org/). Por ejemplo, usamos `sns.lineplot` para hacer gráficos de líneas. Pronto aprenderás que usamos `sns.barplot` y `sns.heatmap` para crear gráficos de barras y mapas de calor, respectivamente._
- `data=spotify_data` selecciona los datos que se usarán para crear el gráfico.

Ten en cuenta que siempre usarás este mismo formato cuando crees un gráfico de líneas, y **_lo único que cambia con un nuevo conjunto de datos es el nombre del dataset_**. Así que, si estuvieras trabajando con un conjunto de datos diferente llamado `financial_data`, por ejemplo, la línea de código sería así:

```
sns.lineplot(data=financial_data)
```

A veces hay detalles adicionales que queremos modificar, como el tamaño de la figura o el título del gráfico. Cada una de estas opciones puede configurarse fácilmente con una sola línea de código.

````python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of each song
sns.lineplot(data=spotify_data)
````

La primera línea de código establece el tamaño de la figura en `14` pulgadas (de ancho) por `6` pulgadas (de alto). Para establecer el tamaño de _cualquier figura_, solo necesitas copiar la misma línea de código tal como aparece. Luego, si quieres usar un tamaño personalizado, cambia los valores `14` y `6` por el ancho y alto deseados.

La segunda línea de código establece el título de la figura. ¡Ten en cuenta que el título *siempre* debe estar entre comillas (`"..."`)!

# Graficar un subconjunto de los datos

Hasta ahora, aprendiste a graficar una línea para _todas_ las columnas del conjunto de datos. En esta sección, aprenderás a graficar un _subconjunto_ de las columnas.

Comenzaremos imprimiendo los nombres de todas las columnas. Esto se hace con una línea de código y puede adaptarse a cualquier conjunto de datos simplemente cambiando el nombre del dataset (en este caso, `spotify_data`).


````python
list(spotify_data.columns)
````

En la siguiente celda de código, graficamos las líneas correspondientes a las dos primeras columnas del conjunto de datos.


````python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")

# Line chart showing daily global streams of 'Despacito'
sns.lineplot(data=spotify_data['Despacito'], label="Despacito")

# Add label for horizontal axis
plt.xlabel("Date")
````

Las dos primeras líneas de código configuran el título y el tamaño de la figura (_¡y deberían resultarte muy familiares!_).

Las siguientes dos líneas agregan una línea al gráfico de líneas. Por ejemplo, considera la primera, que añade la línea para "Shape of You":


```python
# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")
```

Esta línea es muy similar al código que usamos cuando graficamos todas las líneas del conjunto de datos, pero tiene algunas diferencias clave:  
- En lugar de usar `data=spotify_data`, usamos `data=spotify_data['Shape of You']`. En general, para graficar solo una columna, usamos este formato, poniendo el nombre de la columna entre comillas simples y encerrado entre corchetes. (_Para asegurarte de que especificas correctamente el nombre de la columna, puedes imprimir la lista de todas las columnas usando el comando que aprendiste arriba._)  
- También añadimos `label="Shape of You"` para que la línea aparezca en la leyenda con su etiqueta correspondiente.

La última línea de código modifica la etiqueta del eje horizontal (o eje x), donde la etiqueta deseada se coloca entre comillas (`"..."`).
