## Indexación, selección y slicing

Las columnas individuales en el DataFrame se pueden seleccionar con indexación regular. Cada columna está representada como una `Series`:


```python
df
```

```python
df.loc['Canada']
```

```python
df.iloc[-1]
```

```python
df['Population']
```

Notese que el `index` de la Serie devuelta es el mismo que el del DataFrame. Y su `name` es el nombre de la columna. Si estás trabajando en un notebook y quieres ver un formato más parecido a un DataFrame, puedes usar el método `to_frame`:


```python
df['Population'].to_frame()
```

También se pueden seleccionar múltiples columnas de manera similar a `numpy` y a las `Series`:


```python
df[['Population', 'GDP']]
```

En este caso, el resultado es otro `DataFrame`. El slicing funciona de manera diferente: actúa a nivel de "fila" y puede ser un poco contraintuitivo:


```python
df[1:3]
```

La selección a nivel de filas funciona mejor con `loc` e `iloc`, **los cuales se recomiendan** por encima del "slicing directo" (`df[:]`).

`loc` selecciona las filas que coinciden con el índice dado:


```python
df.loc['Italy']
```

```python
df.loc['France': 'Italy']
```

As a second "argument", you can pass the column(s) you'd like to select:

```python
df.loc['France': 'Italy', 'Population']
```

```python
df.loc['France': 'Italy', ['Population', 'GDP']]
```

`iloc` funciona con la "posición" (numérica) del índice:


```python
df
```

```python
df.iloc[0]
```

```python
df.iloc[-1]
```

```python
df.iloc[[0, 1, -1]]
```

```python
df.iloc[1:3]
```

```python
df.iloc[1:3, 3]
```

```python
df.iloc[1:3, [0, 3]]
```

```python
df.iloc[1:3, 1:3]
```

> **RECOMENDADO: Usa siempre `loc` e `iloc` para reducir ambigüedades, especialmente con `DataFrame`s que tienen índices numéricos.**

## Selección condicional (arreglos booleanos)

Vimos la selección condicional aplicada a las `Series` y funcionará de la misma manera para los `DataFrame`s. Después de todo, un `DataFrame` es una colección de `Series`:


```python
df
```

```python
df['Population'] > 70
```

```python
df.loc[df['Population'] > 70]
```

La comparación booleana se realiza a nivel de Índice, por lo que puedes filtrar por cualquier fila, siempre que contenga los índices correctos. La selección de columnas sigue funcionando como se espera:


```python
df.loc[df['Population'] > 70, 'Population']
```

```python
df.loc[df['Population'] > 70, ['Population', 'GDP']]
```

## Eliminar elementos

En contraste con el concepto de selección, tenemos el de "eliminar". En lugar de indicar qué valores te gustaría _seleccionar_, puedes indicar cuáles te gustaría `eliminar`:


```python
df.drop('Canada')
```

```python
df.drop(['Canada', 'Japan'])
```

```python
df.drop(columns=['Population', 'HDI'])
```

```python
df.drop(['Italy', 'Canada'], axis=0)
```

```python
df.drop(['Population', 'HDI'], axis=1)
```

```python
df.drop(['Population', 'HDI'], axis=1)
```

```python
df.drop(['Population', 'HDI'], axis='columns')
```

```python
df.drop(['Canada', 'Germany'], axis='rows')
```

Todos estos métodos `drop` devuelven un nuevo `DataFrame`. Si deseas modificarlo "en el lugar" (in place), puedes usar el atributo `inplace` (hay un ejemplo más abajo).

## Modificando DataFrames

Es simple e intuitivo. Puedes agregar columnas o reemplazar valores en columnas sin problemas:


### Adding a new column

```python
langs = pd.Series(
    ['French', 'German', 'Italian'],
    index=['France', 'Germany', 'Italy'],
    name='Language'
)
```

```python
langs
```

```python
df['Language'] = langs
```

```python
df
```

---
### Reemplazando valores por columna


```python
df['Language'] = 'English'
```

```python
df
```

---
### Renombrando Columnas


```python
df.rename(
    columns={
        'HDI': 'Human Development Index',
        'Anual Popcorn Consumption': 'APC'
    }, index={
        'United States': 'USA',
        'United Kingdom': 'UK',
        'Argentina': 'AR'
    })
```

```python
df.rename(index=str.upper)
```

```python
df.rename(index=lambda x: x.lower())
```

---
### Removiendo Columnas

```python
df.drop(columns='Language', inplace=True)
```

---
### Agregando valores

```python
df.append(pd.Series({
    'Population': 3,
    'GDP': 5
}, name='China'))
```

*Append* returna un nuevo `DataFrame`:

```python
df
```

Puedes asignar directamente el nuevo índice y los valores al `DataFrame`:


```python
df.loc['China'] = pd.Series({'Population': 1_400_000_000, 'Continent': 'Asia'})
```

```python
df
```

Podemos usar `drop` para eliminar una fila por índice:


```python
df.drop('China', inplace=True)
```

```python
df
```

---
### More radical index changes

```python
df.reset_index()
```

```python
df.set_index('Population')
```

## Crear columnas a partir de otras columnas

Modificar un DataFrame a menudo implica combinar diferentes columnas en otra. Por ejemplo, en nuestro análisis de países, podríamos intentar calcular el "PIB per cápita", que es simplemente `PIB / Población`.


```python
df[['Population', 'GDP']]
```

La forma habitual de expresar eso en pandas es simplemente dividiendo cada serie:


```python
df['GDP'] / df['Population']
```

El resultado de esa operación es simplemente otra serie que puedes agregar al `DataFrame` original:


```python
df['GDP Per Capita'] = df['GDP'] / df['Population']
```

```python
df
```
