## Arreglos booleanos en Series (también llamados máscaras)

### Dada la Serie X de pandas, crear una máscara que muestre los elementos negativos

```python
# your code goes here
```

```python
X = pd.Series([-1,2,0,-4,5,6,0,0,-9,10])

mask = X <= 0
mask
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

### Dada la Serie X de pandas, obtener los elementos negativos

```python
# your code goes here
```

```python
X = pd.Series([-1,2,0,-4,5,6,0,0,-9,10])

mask = X <= 0
X[mask]
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

### Dada la Serie X de pandas, obtener números mayores que 5

```python
# your code goes here
```

```python
X = pd.Series([-1,2,0,-4,5,6,0,0,-9,10])

mask = X > 5
X[mask]
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

### Dada la Serie X de pandas, obtener números mayores que la media de sus elementos

```python
# your code goes here
```

```python
X = pd.Series([-1,2,0,-4,5,6,0,0,-9,10])

mask = X > X.mean()
X[mask]
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

### Dada la Serie X de pandas, obtener números iguales a 2 o 10

```python
# your code goes here
```

```python
X = pd.Series([-1,2,0,-4,5,6,0,0,-9,10])

mask = (X == 2) | (X == 10)
X[mask]
```
