# Introducción
Comenzaremos con una visión general de cómo funcionan los modelos de *machine learning* y cómo se utilizan. Esto puede parecer básico si ya has hecho modelado estadístico o trabajado con *machine learning* antes. No te preocupes, pronto avanzaremos hacia la construcción de modelos más potentes.

Esta sección te hará construir modelos a medida que avanzas en el siguiente escenario:

Tu primo ha ganado millones de dólares especulando en bienes raíces. Te ha ofrecido asociarse contigo en los negocios debido a tu interés en la ciencia de datos. Él pondrá el dinero y tú aportarás modelos que predigan cuánto valen distintas casas.

Le preguntas a tu primo cómo ha predicho los valores inmobiliarios en el pasado, y él dice que simplemente por intuición. Pero al insistir un poco más, revela que ha identificado patrones de precios de casas que ha visto antes, y usa esos patrones para hacer predicciones sobre nuevas propiedades que está considerando.

El *machine learning* funciona de la misma manera. Comenzaremos con un modelo llamado **Árbol de Decisión** (*Decision Tree*). Existen modelos más sofisticados que ofrecen predicciones más precisas, pero los árboles de decisión son fáciles de entender y constituyen la base de algunos de los mejores modelos en ciencia de datos.

Por simplicidad, empezaremos con el árbol de decisión más sencillo posible.

![Primeros Árboles de Decisión](https://storage.googleapis.com/kaggle-media/learn/images/7tsb5b1.png)

Este divide las casas en solo dos categorías. El precio predicho para cualquier casa en consideración es el precio promedio histórico de casas en la misma categoría.

Usamos datos para decidir cómo dividir las casas en dos grupos, y luego otra vez para determinar el precio predicho en cada grupo. Este paso de capturar patrones a partir de los datos se llama **ajustar** (*fitting*) o **entrenar** (*training*) el modelo. Los datos utilizados para **ajustar** el modelo se denominan **datos de entrenamiento** (*training data*).  

Los detalles de cómo se ajusta el modelo (por ejemplo, cómo dividir los datos) son lo suficientemente complejos como para dejarlos para más adelante. Una vez que el modelo ha sido ajustado, puedes aplicarlo a nuevos datos para **predecir** (*predict*) los precios de otras viviendas.

# Mejorando el Árbol de Decisión
¿Cuál de los siguientes dos árboles de decisión es más probable que resulte del ajuste a los datos de entrenamiento de bienes raíces?

![Primeros Árboles de Decisión](https://storage.googleapis.com/kaggle-media/learn/images/prAjgku.png)

El árbol de decisión de la izquierda (*Decision Tree 1*) probablemente tiene más sentido, porque capta la realidad de que las casas con más habitaciones tienden a venderse a precios más altos que aquellas con menos habitaciones.

La mayor limitación de este modelo es que no capta la mayoría de los factores que afectan el precio de una vivienda, como el número de baños, el tamaño del terreno, la ubicación, etc.

Puedes capturar más factores usando un árbol con más "divisiones" (*splits*). Estos se llaman árboles más "profundos" (*deeper trees*). Un árbol de decisión que también considera el tamaño total del terreno de cada casa podría verse así:  
![Árbol con profundidad 2](https://storage.googleapis.com/kaggle-media/learn/images/R3ywQsR.png)

Predices el precio de cualquier casa recorriendo el árbol de decisión, eligiendo siempre el camino correspondiente a las características de esa casa. El precio predicho para la casa se encuentra al final del árbol. El punto en el que hacemos la predicción se llama una **hoja** (*leaf*).   

Las divisiones (*splits*) y los valores en las hojas serán determinados por los datos, así que es hora de que revises los datos con los que vas a trabajar.
